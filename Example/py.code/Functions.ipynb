### Exercies- Heron's Formula 헤론의 공식
   


```python
def area_of_triangle(vertices):
    """Compute the area of the triangle with given vertices.

    Parameters
    ----------
    vertices : list of tuples of numbers
        The vertices of a triangle [(x1, y1), (x2, y2), (x3, y3)].

    Returns
    -------
    float
        Area of the triangle computed by Heron's formula.

    Examples
    --------
    >>> area_triangle([(0, 0), (3, 0), (3, 4)])
    6.0
    >>> area_triangle([(-1, 2), (-3, -1), (4, 1)])
    8.499999999999996
    """
    P, Q, R = vertices
    a = sum((x-y) ** 2 for x, y in zip(P,Q)) ** 0.5
    b = sum((x-y) ** 2 for x, y in zip(P,R)) ** 0.5
    c = sum((x-y) ** 2 for x, y in zip(Q, R)) ** 0.5    
    print(zip(P,Q))
    print(zip(P,R))
    s = (a + b + c) / 2
    return(s * (s - a) * (s - b) * (s - c)) ** 0.5
```

### Riemann Zeta Function

```python
def zeta(s, N):
    return sum(i ** -s for i in range(1, N + 1))
```

```python
zeta(1, 1) == 1.0
```

```python
zeta(2, 2) == 2.5/2
```

```python
[zeta(s, 100000) for s in (2,4)]
```

### 기하 평균

#### 변화율에 대한 평균값을 구할 때

### 조화 평균 
#### 속도를 구할때 많이 씀

```python
def harmonic_mean(seq) :
    return len(seq) / sum(1/ x for x in seq)
```

```python
harmonic_mean((1, 4, 4))
```

```python
harmonic_mean([1, 1, 1, 1])
```

```python
harmonic_mean(range(1, 4))
```
