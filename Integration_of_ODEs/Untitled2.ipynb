```python
import numpy as np
from matplotlib import pyplot as plt
plt.rcParams.update({'font.size': 15})
#%config InlineBackend.figure_format='retina'
```

```python
k = 0.3

def T(w):     # K.E.
    return 0.5 * w*w

def V(th):    # P.E.
    return -np.cos(th) + k*th

def H(w, th): # Hamiltonian a.k.a total energy
    return T(w) + V(th)

th_min = np.arcsin(-k)   # theta at P.E. min
V_min  = V(th_min)       # P.E. min
th_sep = -np.pi - th_min # theta at the saddle point of separatrix
V_sep  = V(th_sep)       # P.E. at the saddle point of separatrix

# prepare for inputs to pyplot.contour
x = np.linspace(-np.pi, np.pi, 201) # theta domain
y = np.linspace(-2, 2, 200)         # omega domain

xx, yy = np.meshgrid(x, y)
zz = H(yy, xx)

levels = np.hstack([                # contour levels
    np.linspace(V_min, V_sep, 7)[1:],
    np.linspace(0, 3, 7)[1:] + V_sep
])

# draw
plt.figure()

plt.contour(x, y, zz, levels, linewidths = 1)
plt.contour(x, y, zz, [V_sep], colors="k")
plt.plot(th_min + 0*y, y, "-.m", lw = 1)
plt.plot(th_sep + 0*y, y, "--r", lw = 1)

plt.xlabel("$\\theta$")
plt.ylabel("$\\omega$")

plt.show()
```
