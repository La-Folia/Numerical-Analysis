```python
import numpy as np
from scipy import fft
from matplotlib import pyplot as plt
plt.rcParams.update({'font.size': 15})
#%config InlineBackend.figure_format='retina'
```

```python
N  = 10
L  = 2*np.pi

xj = np.linspace(0, L, N + 1)[:-1]
uj = np.exp(2j*xj)

Ul = fft.fft(uj) # this time using scipy.fft.fft
kl = np.hstack([
    np.arange(0, N/2),       # k > 0 domain
    np.arange(-int(N/2), 0), # k < 0 domain
]) * 2*np.pi/L

# sort in ascending order of k
i, kl = zip(*sorted(enumerate(kl), key=lambda x: x[1]))
Ul = Ul[i,]

plt.figure()
plt.plot(kl, Ul.real, "or", kl, Ul.imag, "^b")
plt.legend(["real", "imag"])
plt.xlabel("$k$")
plt.ylabel("$\\hat{u}(k)$")
plt.grid()
plt.show()
```

```python
# Parseval's theorem
lhs = sum(abs(uj)**2)
rhs = np.mean(abs(Ul)**2)
print(f"""
lhs = {lhs}, rhs = {rhs}
""")
```

```python
# Check if IDFT gives the original data
iUl = fft.ifft(Ul[i,])

print(uj - iUl)
```

```python
note = [
     "C3",  "E3",  "G3",  "C4",  "E4",   "G4",   "D5",   "E5",   "G5",
     "B5",  "D6",  "F6",  "G6", "A#6",   "B6",   "D7",   "E7",
]
frequency = [ # in Hz
    132.3, 165.4, 198.4, 264.6,   333,  394.7,  593.1,  663.7,   796, 
    996.7,  1186,  1387,  1592,  1857,   1991,   2324,   2657
]
amplitude = [
    13.28, 9.081, 223.6, 16.71, 30.98,  13.28,  2.633,  6.258, 2.815, 
    46.93, 1.125, 1.797, 1.458, 2.295, 0.7868, 0.5319, 0.6912
]

plt.figure(figsize=[30, 3])
ax = plt.subplot()
plt.loglog(frequency, amplitude, "or")
for x, y in zip(frequency, amplitude):
    plt.loglog([x, x], [0, y], ":r")
plt.xlabel("note")
plt.ylabel("amplitude")
plt.xticks(frequency, note)
ax.tick_params(axis='x', which='minor', bottom=False)
plt.show()
```

```python
# number of samples per second, i.e., 1/Δt
sampling_frequency = 48000 // 8

# then, maximum frequency that can be resolved
max_freq = 0.5 * sampling_frequency

# discrete time domain
max_time = 2 # in seconds
time = np.linspace(0, max_time, sampling_frequency * 2 + 1)[:-1]

# random phases for different notes
phase = np.random.rand(len(time)) * 2*np.pi

# wave samples
wave  = sum(A * np.cos(2*np.pi*f*time + ph) 
            for A, f, ph in zip(amplitude, frequency, phase))
```

```python
Wj = fft.fft(wave)
N  = len(Wj)

Dk = 2*np.pi / (max(time) - min(time))

kj = np.hstack([
    np.arange(0,   N/2),     # k > 0 domain
    np.arange(-int(N/2), 0), # k < 0 domain
]) * Dk #12000

# kj is 'angular' frequency.
# divide it by 2pi to convert to frequency in Hz
fj = kj / (2*np.pi)

# power spectral density
Pj = np.abs(Wj/N)**2 / Dk

plt.figure(figsize=[11, 3])
ax = plt.subplot()
plt.loglog(fj[:int(N/2)], Pj[:int(N/2)], "-r", lw=1)
plt.xlim([frequency[0]*0.8, frequency[-1]*1.2])
plt.xlabel("note")
plt.ylabel("$P(k)$")
plt.xticks(frequency, note)
ax.tick_params(axis='x', which='minor', bottom=False)
plt.grid(axis="x")
plt.show()


```

```python
# find the index to "G4" note
index = [i for i, n in enumerate(note) if n == "G4"][0]


# draw a line in between "G4" and "D5"
freq_cut = 0.5 * (frequency[index] + frequency[index + 1])

# find indices of the frequencies higher than cut-off freq
indices = np.abs(fj) > freq_cut

# decimate wave amplitude
filtered_Wj = Wj.copy()
filtered_Wj[indices,] *= 0

# inverse FFT
filtered_wave = fft.ifft(filtered_Wj).real # since we zero-ed out symmetrically, 
                                           # we are sure that ifft result is real
```
