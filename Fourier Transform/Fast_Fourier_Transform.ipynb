```python
import numpy as np
from scipy import fft
from matplotlib import pyplot as plt
plt.rcParams.update({'font.size': 15})
#%config InlineBackend.figure_format='retina'
```

```python
N  = 10
L  = 2*np.pi

xj = np.linspace(0, L, N + 1)[:-1]
uj = np.exp(2j*xj)

Ul = fft.fft(uj) # this time using scipy.fft.fft
kl = np.hstack([
    np.arange(0, N/2),       # k > 0 domain
    np.arange(-int(N/2), 0), # k < 0 domain
]) * 2*np.pi/L

# sort in ascending order of k
i, kl = zip(*sorted(enumerate(kl), key=lambda x: x[1]))
Ul = Ul[i,]

plt.figure()
plt.plot(kl, Ul.real, "or", kl, Ul.imag, "^b")
plt.legend(["real", "imag"])
plt.xlabel("$k$")
plt.ylabel("$\\hat{u}(k)$")
plt.grid()
plt.show()
```

```python
# Parseval's theorem
lhs = sum(abs(uj)**2)
rhs = np.mean(abs(Ul)**2)
print(f"""
lhs = {lhs}, rhs = {rhs}
""")
```

```python
# Check if IDFT gives the original data
iUl = fft.ifft(Ul[i,])

print(uj - iUl)
```
